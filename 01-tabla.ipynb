Genera una tabla cargando del excel los datos iniciales de las imágenes, y añadiendo las medidas realizadas por nosotros

```python
import pandas as pd
import numpy as np
import glob
import re
import re
```

```python
tablepath='datos/medidas.xlsx'
# Echar vistazo a esto https://en.wikipedia.org/wiki/Student%27s_t-distribution#Table_of_selected_values
```

```python
lambdadisco= {'Halfa-7 nm':656.281,
              'Halfa-35 nm':656.281,
              'Hbeta-8.5nm':486.135}
dlambdadisco={'Halfa-7 nm':7,
              'Halfa-35 nm':35,
              'Hbeta-8.5nm':8.5}
alphap_mas  ={'Sirio':6.04,
              'Rigel':2.77,
              'Betelgeuse':54.04,
              'GammaLeo':7.70}
filtername= {'Halfa-7 nm':'H-alpha (7 nm)',
              'Halfa-35 nm':'H-alpha (35 nm)',
              'Hbeta-8.5nm':'H-beta (8.5 nm)'}
starname={'Sirio':'Sirius',
          'Rigel':'Rigel',
          'Betelgeuse':'Betelgeuse',
          'GammaLeo':'GammaLeo'}

alphap={k:ap*(np.pi/180/1000/3600) for k, ap in alphap_mas.items()} #rad
```

```python
dts={}#{'D (mm)':np.int16,'L (mm)':np.int16,'L1 (mm)':np.int16,'L2 (mm)':np.int16}
cols=['D (mm)','L (mm)','L1 (mm)','L2 (mm)']
df=pd.read_excel(tablepath,engine="openpyxl",index_col=0,converters=dts                 
                )
df
```

```python
df.dtypes
```

```python
for c in cols:
    df[c]=pd.to_numeric(df[c],errors='coerce')
df['Filtro']=df['Filtro'].str.strip()
df['Objeto']=df['Objeto'].str.strip()
df['Disco']=df['Disco'].str.strip()
```

```python
df.dtypes
```

```python
names=[]
ims=[]
for name in glob.glob('datos/imagenes/*/*.tif'):
    names.append(name)
    ims.append(int(re.search('\d+', name.split('\\')[-1])[0]))

paths=pd.DataFrame({'PATH':names,'Imagen':ims},index=ims)
paths
```

```python
df2=df.join(paths)
df2
```

```python
df2['$\lambda$']=df2['Filtro'].map(lambdadisco)
df2['$d\lambda$']=df2['Filtro'].map(dlambdadisco)
df2['alphap']=df2['Objeto'].map(alphap)
df2['alphap_mas']=df2['alphap'].apply(lambda x: x/ (np.pi/180/1000/3600))
df2['filtername']=df2['Filtro'].map(filtername)
df2['starname']=df2['Objeto'].map(starname)
```

```python
# df2['V_exp']=df2['Objeto'].map(starname)

# s='1Sirio0005 Bx3 Halfa 7 nm Inter1_I_D22mm-L220mm_2.4s Gmax 17-03-10 20-48-51.tif'
# re.findall("\d{2}-\d{2}-\d{2}", s.split('\\')[-1])[0]
```

```python
df2['DATE']=df2['PATH'].apply(lambda x: re.findall("\d{2}-\d{2}-\d{2}", x.split('\\')[-1])[0])
df2['DAY']=df2['DATE'].apply(lambda x:int(x.split('-')[-1]))
```

```python
df2['HOUR']=df2['PATH'].apply(lambda x: re.findall("\d{2}-\d{2}-\d{2}", x.split('\\')[-1])[1].replace('-',':'))

```
